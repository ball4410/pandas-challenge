```python
#Modules
import os
import pandas as pd

#Path to data files
schools_path = os.path.join('.', "schools_complete.csv")
students_path = os.path.join('.', "students_complete.csv")
```

```python
# Use Pandas to read files
schools_df = pd.read_csv(schools_path)
schools_df.head()

students_df = pd.read_csv(students_path)
#students_df.head()
```

```python
#Find total schools
schools_df["school_name"]
len(schools_df["school_name"])
total_schools = len(schools_df["school_name"])
#total_schools
```

```python
#Find total students
schools_df["size"]
schools_df["size"].sum()
total_students = schools_df["size"].sum()
#total_students
```

```python
#Find total budget
schools_df["budget"]
schools_df["budget"].sum()
total_budget = schools_df["budget"].sum()
#total_budget
```

```python
#Find average math score
students_df["math_score"]
students_df["math_score"].mean()
average_math_score = students_df["math_score"].mean()
#average_math_score
```

```python
#Find average reading score
students_df["reading_score"]
students_df["reading_score"].mean()
average_reading_score = students_df["reading_score"].mean()
#average_reading_score
```

```python
#Find percent of passing math scores
#Passing math
students_df[students_df["math_score"] > 69].head() 
passing_math = len(students_df[students_df["math_score"] > 69])
passing_math/total_students
percent_passing_math = passing_math/total_students
#percent_passing_math
```

```python
#Find percent of passing reading scores
#Passing reading
students_df[students_df["reading_score"] > 69].head() 
len(students_df[students_df["reading_score"] > 69])
passing_reading = len(students_df[students_df["reading_score"] > 69])
passing_reading/total_students
percent_passing_reading = passing_reading/total_students
#percent_passing_reading
```

```python
#Find percent of passing math and reading scores
#Passing math and reading
students_df.loc[(students_df["reading_score"] > 69) & (students_df["math_score"] > 69), :]
len(students_df.loc[(students_df["reading_score"] > 69) & (students_df["math_score"] > 69), :])
passing_both = len(students_df.loc[(students_df["reading_score"] > 69) & (students_df["math_score"] > 69), :])
#passing_both/total_students
#percent_passing_both = passing_both/total_students
```

```python
# Place all of the data found into a summary DataFrame
district_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                              "Total Students": total_students,
                              "Total Budget": total_budget,
                              "Average Math Score": average_math_score,
                              "Average Reading Score": average_reading_score,
                              "Percent Passing Math": percent_passing_math,
                              "Percent Passing Reading": percent_passing_reading,
                              "Percent Passing Math and Reading": percent_passing_both})
district_summary_df
```

```python
#Find per student budget for each school
schools_df.groupby(["school_name"]).sum()
grouped_schools = schools_df.groupby(["school_name"])
grouped_schools

school_averages_df = pd.DataFrame(grouped_schools["size", "budget"].sum())
schools_df["budget"]/schools_df["size"]
per_student_budget = schools_df["budget"]/schools_df["size"]
per_student_budget

school_averages_df["Budget per Student"] = school_averages_df["budget"]/school_averages_df["size"]
#school_averages_df
```

```python
#Find average math and reading scores for each school
students_df.head()
grouped_schools = students_df.groupby('school_name')
grouped_schools
average_scores_per_school_df = pd.DataFrame(grouped_schools["reading_score", "math_score"].mean())
#average_scores_per_school_df

```

```python
#Find percent of students passing math for each school
students_df.head()
grouped_schools.count().head()

students_passing_math = students_df[students_df["math_score"] > 69]
students_passing_math
students_passing_math.groupby(['school_name']).count()
student_passing_math_grouped_by_school = students_passing_math.groupby(['school_name']).count()
students_passing_math_df = pd.DataFrame(student_passing_math_grouped_by_school["math_score"])
students_passing_math_df

grouped_schools["math_score"].count()
students_passing_math_df["math_score"]/grouped_schools["math_score"].count()
students_passing_math_df["Percent Passing Math"] = students_passing_math_df["math_score"]/grouped_schools["math_score"].count()
#students_passing_math_df
```

```python
#Find percent of students passing reading for each school
students_df.head()
grouped_schools.count().head()

students_passing_reading = students_df[students_df["reading_score"] > 69]
students_passing_reading
students_passing_reading.groupby(['school_name']).count()
student_passing_reading_grouped_by_school = students_passing_reading.groupby(['school_name']).count()
students_passing_reading_df = pd.DataFrame(student_passing_reading_grouped_by_school["reading_score"])
students_passing_reading_df

grouped_schools["reading_score"].count()
students_passing_reading_df["reading_score"]/grouped_schools["reading_score"].count()
students_passing_reading_df["Percent Passing Reading"] = students_passing_reading_df["reading_score"]/grouped_schools["reading_score"].count()
#students_passing_reading_df
```

```python
#Find percent of students passing math and reading for each school
students_df.head()
grouped_schools.count().head()

students_df.loc[(students_df["reading_score"] > 69) & (students_df["math_score"] > 69), :]
students_passing_both = students_df.loc[(students_df["reading_score"] > 69) & (students_df["math_score"] > 69), :]
students_passing_both
students_passing_both.groupby(['school_name']).count()
student_passing_both_grouped_by_school = students_passing_both.groupby(['school_name']).count()
student_passing_both_grouped_by_school
students_passing_both_df = pd.DataFrame(student_passing_both_grouped_by_school["reading_score"])
students_passing_both_df

grouped_schools["reading_score"].count()
students_passing_both_df["reading_score"]/grouped_schools["reading_score"].count()
students_passing_both_df["Percent Passing Math and Reading"] = students_passing_both_df["reading_score"]/grouped_schools["reading_score"].count()
#students_passing_both_df
```

```python
#reset index to combine grouped by data 
schools_df
indexed_schools = schools_df.set_index("school_name")

# Place all of the data found into a summary DataFrame
school_summary_df = pd.DataFrame({ #"School Name": indexed_schools["school_name"],
                                  "School Type": indexed_schools["type"],
                                  "Total Students": indexed_schools["size"],
                                  "Total School Budget": indexed_schools["budget"],
                                  "Budget per Student": school_averages_df["Budget per Student"],
                                  "Average Math Score": average_scores_per_school_df["math_score"],
                                  "Average Reading Score": average_scores_per_school_df["reading_score"],
                                  "Percent Passing Math": students_passing_math_df["Percent Passing Math"],
                                  "Percent Passing Reading":students_passing_reading_df["Percent Passing Reading"],
                                  "Percent Passing Math and Reading": students_passing_both_df["Percent Passing Math and Reading"]})

school_summary_df

```

```python
#Find top 5 performing schools based on % Overall Passing
top_performing = school_summary_df.sort_values(["Percent Passing Math and Reading"], ascending=False)
top_performing_df = top_performing.head()
top_performing_df
```

```python
#Find bottom 5 performing schools based on % Overall Passing
bottom_performing = school_summary_df.sort_values(["Percent Passing Math and Reading"], ascending=True)
bottom_performing_df = bottom_performing.head()
bottom_performing_df
```

```python
#Find average Math Score for students of each grade level 
students_df.groupby(['school_name', 'grade'])
grouped_students_df = students_df.groupby(['school_name', 'grade'])
grouped_students_df['math_score'].mean()
average_math_score_per_grade = grouped_students_df['math_score'].mean()
average_math_score_per_grade
```

```python
#Find average Reading Score for students of each grade level 
grouped_students_df['reading_score'].mean()
average_reading_score_per_grade = grouped_students_df['reading_score'].mean()
average_reading_score_per_grade
```

```python
#Create a table that breaks down school performances based on average Spending Ranges 
school_summary_df['Budget per Student'].min() #578
school_summary_df['Budget per Student'].max() #655

bins = [575, 597, 618, 639, 660]

group_names = ["Lowest", "Moderately Low", "Moderately High", "Highest"]

pd.cut(school_summary_df["Budget per Student"], bins, labels=group_names, include_lowest=True)

scores = school_summary_df[['Budget per Student','Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Percent Passing Math and Reading']]
scores 

scores["Spending"] = pd.cut(scores["Budget per Student"], bins, labels=group_names, include_lowest=True)
scores

school_spending_df = scores[['Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Percent Passing Math and Reading', 'Spending']]
school_spending_df
```

```python
#Create a table that breaks down school performances based on size
school_summary_df['Total Students'].min() #427
school_summary_df['Total Students'].max() #4976

bins = [425, 1950, 3475, 5000]

group_names = ["Small", "Medium", "Large"]

pd.cut(school_summary_df["Total Students"], bins, labels=group_names, include_lowest=True)

size = school_summary_df[['Total Students','Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Percent Passing Math and Reading']]
size

size["School Size"] = pd.cut(size["Total Students"], bins, labels=group_names, include_lowest=True)
size

school_size_df = size[['Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Percent Passing Math and Reading', 'School Size']]
school_size_df
```

```python
#Create a table that breaks down school performances based on school type
school_summary_df.groupby(['School Type']).mean()
school_type = school_summary_df.groupby(['School Type'])

school_type_df = school_type[['Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Percent Passing Math and Reading']].mean()
school_type_df
```
